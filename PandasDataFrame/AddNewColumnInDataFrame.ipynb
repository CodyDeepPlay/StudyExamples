## Operation on Pandas Data Frame

First read the example data file, temps.csv

```python
import pandas as pd
# Read in data and display first 5 rows
df = pd.read_csv('temps.csv')
df.head(5)
```

### Add a new column in the data frame, using 'lambda' function key word in python

Using pandas dataframe's apply function, the value of new column will be based on the calculation of values in the current column in the data frame.

```python
# add a new column called "error_percent"
df['error_percent'] = df.apply(lambda row: (row.actual - row.average )/row.average, axis = 1) 
df.head(6)
```

### Add a new column in the data frame, using a custom function
In this example, we will define a custom function, and then feed it into the dataframe creatation process.

```python
def getPercent(actual, average):
    myPercent = (actual-average)/average
    return myPercent

df['error_percent2'] = df.apply(lambda row: getPercent(row.actual, row.average), axis = 1) 
df.head(6)
```
